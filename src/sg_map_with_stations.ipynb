```python
#!conda install -c conda-forge folium=0.5.0 --yes
import folium
```

```python
# define the world map centered around Canada with a low zoom level
sg_map = folium.Map(location=[1.29587,103.82], zoom_start=11)

# display Singapore map
sg_map
```

## Read stations locations

```python
stations = pd.read_csv('../res/stations.csv')
```

```python
stations.head(10)
```

## Plot on map

```python
stations.type.unique()
```

```python
colors = ['yellow','blue','glyphicon', 'red','fa']
all_types = ['air_temp', 'hymidity', 'air', 'rainfall', 'wind']
```

```python
for t, color in zip(all_types, colors):
    df_station = stations[stations.type ==t]
    for lat, lng, in zip(df_station.latitude, df_station.longitude):
        sg_map.add_child(
            folium.CircleMarker(
                [lat, lng],
                radius=5, # define how big you want the circle markers to be
                #color=color,
                fill=True,
                fill_color=color,
                fill_opacity=0.6,
                popup=t
            )
        )
```
