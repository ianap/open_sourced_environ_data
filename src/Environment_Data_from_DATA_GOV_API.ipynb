# Daily Temperature

```python
#generate a list of dates:
sdate=pd.to_datetime('2021-05-01').date()
edate=pd.to_datetime('2021-05-03').date()
DATES1 = pd.date_range(sdate,edate-timedelta(days=1),freq='d')
DATES1=[date_obj.strftime('%Y-%m-%d') for date_obj in DATES1]
df_daily_temp = pd.DataFrame(columns=["date", "time","temp"])
```

```python
for DATE in DATES1:
    url = 'https://api.data.gov.sg/v1/environment/air-temperature?&date={}'.format(DATE)
    results = requests.get(url).json()
    multiple_level_data = json_normalize(data=results['items'], record_path='readings', meta=['timestamp'])
    if results['items']:
        multiple_level_data[['date','time']] = multiple_level_data.timestamp.str.split("T",expand=True) 
        multiple_level_data.columns = ["station_id", "temp", "timestamp","date", "time"]
        row = multiple_level_data[["date", "time","station_id","temp"]]
        df_daily_temp = df_daily_temp.append(row)
```

## Daily Humidity

```python
#generate a list of dates:
sdate=pd.to_datetime('2021-05-01').date()
edate=pd.to_datetime('2021-05-02').date()
DATES1 = pd.date_range(sdate,edate-timedelta(days=1),freq='d')
DATES1=[date_obj.strftime('%Y-%m-%d') for date_obj in DATES1]
df_daily_humidity = pd.DataFrame(columns=["date","humidity"])
```

```python
for DATE in DATES1:
    print(DATE)
    url = 'https://api.data.gov.sg/v1/environment/relative-humidity?&date={}'.format(DATE)
    results = requests.get(url).json()
    if results['items']:
        if results['items'][0]['readings']:
            multiple_level_data = json_normalize(data=results['items'], record_path='readings', meta=['timestamp'])

            multiple_level_data[['date','time']] = multiple_level_data.timestamp.str.split("T",expand=True) 
            multiple_level_data.columns = ["station_id", "humidity", "timestamp","date", "time"]
            row = multiple_level_data[["date", "time",'station_id',"humidity"]]
            row2 = row.groupby(['date']).mean()['humidity'].reset_index()
            df_daily_humidity = df_daily_humidity.append(row)
```
